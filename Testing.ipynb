```python
from PIL import Image
from keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
```

```python
def image_to_number(model, image_path, display = True):
    img = Image.open(image_path).convert('L')  # convert image to 8-bit grayscale
    WIDTH, HEIGHT = img.size
    data = list(img.getdata()) # convert image data to a list of integers
    # convert that to 2D list (list of lists of integers)
    data = [data[offset:offset+WIDTH] for offset in range(0, WIDTH*HEIGHT, WIDTH)]
    reshaped_data = np.array(data).reshape(1,784)
    number_guessed = model.predict_classes(reshaped_data)[0]
    if display:
        plt.imshow(data, cmap='gray', interpolation='none')
        plt.title("Class {}".format(str(number_guessed)))
    return number_guessed
```
